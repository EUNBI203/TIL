```python
import random
class Pikachu:
    
    def __init__(self, name):
        self.name = name
        self.level = 5
        self.hp = self.level * random.randint(15, 20)
        self.exp = 0
        
    def bark(self):
        return print(f'{self.name}: pikachu!!!!!')
    
    def body_attack(self, someone):
        
        if self.hp <= 0:
            print(f'{self.name} hp가 부족합니다.')
            return False
        if someone.hp <=0:
            print(f'{someone.name} hp가 부족합니다.')
            return False
        
        if random.randint(0, 1):
            someone.hp -= self.level * 5
            print(f'{self.name}가 {someone.name}에게 body_attack !!')
            print(f'{someone.name} hp가 {self.level * 5}만큼 줄었습니다.')
            if someone.hp <= 0:
                someone.hp = 0
            print(f'{self.name} hp: {self.hp}, {someone.name} hp: {someone.hp}')
            print()
            
            if someone.hp <= 0:
                self.exp += someone.level * 15
                print(f'{self.name} win! {self.name}가 경험치 {someone.level * 15}를 얻었습니다.')
                print(f'{self.name} exp: {self.exp}, {someone.name} exp: {someone.exp}')
                print()
                
                if self.exp >= self.level * 100:
                    self.level += 1
                    self.exp = 0
                    print(f'{self.name} 레벨 업!')
                    print(f'{self.name} level: {self.level}, {someone.name} level: {someone.level}')
                    print()
                    return False
                return False
                    
        else:
            print(f'{self.name} 공격이 실패했습니다.')
            print()
        return True
        
    def thousond_volt(self, someone):
        if self.hp <= 0:
            print(f'{self.name} hp가 부족합니다.')
            return False
        if someone.hp <=0:
            print(f'{someone.name} hp가 부족합니다.')
            return False
        
        if random.randint(0, 1):
            someone.hp -= self.level * 7
            print(f'{self.name}가 {someone.name}에게 thousond_volt !!')
            print(f'{someone.name} hp가 {self.level * 7}만큼 줄었습니다.')
            if someone.hp <= 0:
                someone.hp = 0
            print(f'{self.name} hp: {self.hp}, {someone.name}의 hp: {someone.hp}')
            print()
            
            if someone.hp <= 0:
                self.exp += someone.level * 15
                print(f'{self.name} 승리! {self.name}가 경험치 {someone.level * 15}를 얻었습니다.')
                print(f'{self.name} exp: {self.exp}, {someone.name} exp: {someone.exp}')
                print()
                
                if self.exp >= self.level * 100:
                    self.level += 1
                    self.exp = 0
                    print(f'{self.name} 레벨 업!')
                    print(f'{self.name} level: {self.level}, {someone.name} level: {someone.level}')
                    print()
                    return False
                return False
                    
        else:
            print(f'{self.name} 공격이 실패했습니다.')
            print()
        return True
    
    def charge(self):
        self.hp = self.level * random.randint(15, 20)
        return print(f'{self.name} hp가 {self.hp}만큼 충전되었습니다.')
```

```python
jiwoo = Pikachu('지우의 피카츄')
tak = Pikachu('타키의 피카츄')
```

```python
def fight(p1, p2):
    result = True
    while result:
        if random.randint(0, 1):
            p1.bark()
            if random.randint(0, 1):
                result = p1.thousond_volt(p2)
            else:
                result = p1.body_attack(p2)
        else:
            p2.bark()
            if random.randint(0, 1):
                result = p2.thousond_volt(p1)
            else:
                result = p2.body_attack(p1)
```

```python
fight(jiwoo, tak)
```

```python
jiwoo.charge()
```
